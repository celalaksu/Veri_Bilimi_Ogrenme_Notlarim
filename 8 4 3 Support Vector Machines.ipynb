# DESTEM VEKTÖR MAKİNELERİ - SUPPORT VECTOR MECHINES - SVM

```python
import numpy as np 
import pandas as pd 
import statsmodels.api as sm 
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.preprocessing import scale, StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import confusion_matrix, accuracy_score, mean_squared_error, r2_score, roc_auc_score, roc_curve, classification_report 
```

```python
from sklearn.svm import SVC
```

```python
df = pd.read_csv("verisetleri\diabetes.csv")
df.head()
```

```python
y = df["Outcome"]
X = df.drop(["Outcome"], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
# SVM de doğrusal ve doğrusal olamayan birbirinden farklı yaklaşımlar söz konusudur.
# MODEL VE TAHMİN bölümünde doğrusal bir yapı ele alacağız.
# MODEL TUNNING bölümünde ise hem RBF adı verilen hem de LINEAR olan iki farklı yaklaşımı değerlendirmiş olacağız. Yani doğrusal ve doğrusal olmayan SVM i MODEL TUNNING bölümünde ele alacağız.
```

```python
svm_model = SVC().fit(X_train, y_train)
```

```python
svm_model
```

```python
# PARAMETRELER
# kernel = 'rbf' --> Doğrusal olmayan form öntanımlıdır.
```

```python
svm_model = SVC(kernel = "linear").fit(X_train, y_train)
```

```python
y_pred = svm_model.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
# MODEL TUNNING
```

```python
svm = SVC()
```

```python
svm_params = {"C" : np.arange(1, 10), "kernel" : ["linear", "rbf"]}
```

```python
svm_cv_model = GridSearchCV(svm, svm_params, cv=5, n_jobs=-1, verbose=2).fit(X_train, y_train)
```

```python
svm_cv_model.best_score_
```

```python
svm_cv_model.best_params_
```

```python
# FİNAL MODEL
```

```python
# C --> Karmaşıklık, ceza parametresi
svm_tuned = SVC(C=5, kernel="linear").fit(X_train, y_train)
```

```python
y_pred = svm_tuned.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```
